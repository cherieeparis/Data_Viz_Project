```r
library(shiny)
library(DT)
library(dplyr)
library(lubridate)
library(readr)
library(ggplot2)
library(plotly)
library(tidyr)
library(zoo)
library(stringr)
```

```r
# only run if didnt connect to colab drive file
pathname <- '/content/drive/My Drive/Data viz final project/LCA_Data/LCA2019-2023_cleaned_final.csv'
lca <- read.csv(pathname)
```

```r
# clean job title

lca$JOB_TITLE <- as.character(lca$JOB_TITLE)
lca$JOB_TITLE <- gsub("\\s*-.*|\\s*kbg.*|\\s*kboey.*|[0-9]+\\.?[0-9]*|\\[.*?\\]|\\(.*?\\)|[.,]|\\(|\\)|\\[|\\]|\\b(?:all other)\\b", "", lca$JOB_TITLE, perl = TRUE)


# Creating a cleaned title column with further refined cleaning
lca <- lca %>%
  mutate(JOB_TITLE = str_trim(str_remove_all(JOB_TITLE, "[^A-Za-z ]+")))
```

```r
# clean plural words
singularize <- function(title) {

  title <- str_replace_all(title, "(?<!s)ies$", "y")
  title <- str_replace_all(title, "(?<=[aotl])es$", "")
  title <- str_replace_all(title, "(?<!s)s$", "")

  return(title)
}

lca$JOB_TITLE <- sapply(lca$JOB_TITLE, singularize, USE.NAMES = FALSE)
lca$JOB_TITLE <- as.character(lca$JOB_TITLE)
```

```r
lca %>%
  distinct(JOB_TITLE) %>%
  pull(JOB_TITLE) %>%
  print()
```

```r
# clean SOC title, followed same logic

lca$SOC_TITLE <- as.character(lca$SOC_TITLE)
lca$SOC_TITLE <- gsub("\\s*-.*|\\s*kbg.*|\\s*kboey.*|[0-9]+\\.?[0-9]*|\\[.*?\\]|\\(.*?\\)|[.,]|\\(|\\)|\\[|\\]|\\b(?:all other)\\b|\\b(?:all others)\\b", "", lca$SOC_TITLE, perl = TRUE)

lca <- lca %>%
  mutate(SOC_TITLE = str_trim(str_remove_all(SOC_TITLE, "[^A-Za-z ]+")))
```

```r
lca
```

```r

# Apply this function to the SOC_TITLE column
lca$SOC_TITLE <- sapply(lca$SOC_TITLE, singularize, USE.NAMES = FALSE)
lca$SOC_TITLE <- as.character(lca$SOC_TITLE)
```

```r
lca %>%
  distinct(SOC_TITLE) %>%
  pull(SOC_TITLE) %>%
  print()
```
